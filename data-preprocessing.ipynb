```python
# assign directory
directory = './recordings/recordings'
 
# iterate over files in
# that directory
for filename in os.listdir(directory):
    out = filename[:-4]
    
    f = os.path.join(directory, filename)
    print(f)
    
    #subprocess.call(['ffmpeg', '-i', f, './recordings/output/' + out + ".wav"])
    
    #import pydub
    sound = pydub.AudioSegment.from_mp3(f)
    sound.export("./recordings/output/" + out + ".wav", format="wav")
    
```
