```python
# assign directory
directory = './recordings'
 
# iterate over files in that directory
for filename in os.listdir(directory):
    out = filename[:-4]
    
    f = os.path.join(directory, filename)
    print(f)

    sound = pydub.AudioSegment.from_mp3(f)
    sound.export("./output/" + out + ".wav", format="wav")
    
    
```
