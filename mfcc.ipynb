```python
# assign directory
directory = './recordings/output'
mfccss = []
 
# iterate over files in
# that directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    x, sr = librosa.load(f)
    
    #idk how many mfcc we want, default is 20
    mfccs = librosa.feature.mfcc(x, sr=sr)
    print(mfccs.shape)
    mfccss.append(mfccs)
    
```
